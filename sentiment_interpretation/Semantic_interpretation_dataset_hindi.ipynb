```python
from discopy import Ty, Word # Import the classes for the type of word and for the word

s, n = Ty('s'), Ty('n') # Define the types s (sentence) and n (noun)

# Define the words (notice that we include both meaning and grammar)

# nouns (subjects)
man = Word('man', n)
aadamee = Word('aadamee', n)

# nouns (objects)
carpet, cloth, toy = Word('carpet', n), Word('cloth', n), Word('toy', n)
kaaleen, kapada, khilauna = Word('kaaleen', n.r @ n), Word('kapada', n.r @ n), Word('khilauna', n.r @ n)

# adjectives
old, new, red, blue = Word('old', n @ n.l), Word('new', n @ n.l), Word('red', n @ n.l), Word('blue', n @ n.l)
purana, naya, laal, neela = Word('purana', n.r @ n  ), Word('naya', n.r @ n ), Word('laal', n.r @ n ), Word('neela',n.r @ n )

# Transitive verbs
buys, steals, makes  = Word('buys', n.r @ s @ n.l),  Word('steals', n.r @ s @ n.l), Word('makes', n.r @ s @ n.l)
khareedata_hai, churata_hai, banata_hai = Word('khareedata hai',  n.r @ s), Word('churata hai',  n.r @ s ), Word('banata hai', n.r @ s )


```

```python
from discopy import Cup, Id, pregroup
```

### Let us generate string diagram for grammar type 1 -<br>
### Subject + Verb + Object (English)<br>
### Subject + Object + Verb (Hindi)

## S1.

```python
grammar =  Cup(n, n.r) @ Id(s) @ Cup(n.l, n)
sentence = man @ buys @ toy  >> grammar
pregroup.draw(sentence)
```

```python
grammar =  Cup(n,n.r) @ Cup(n,n.r)  @ Id(s)
sentence = aadamee  @ khilauna @ khareedata_hai >> grammar
pregroup.draw(sentence)
```
